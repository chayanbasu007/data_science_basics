### Linear Algebra Operations 

##### 1. Condition Number of a Matrix

***
The condition number measures how sensitive a system of linear equations Ax=B (imagine linear regression) is to small changes or numerical rounding errors. Basically assess the numerical stability.
- If the condition number is low then it is good, small changes in the input does not change the output by much - close to 1 is good as there is computation statbility.
- If the condition number is high then it is a ill-conditioned matrix, small changes or tweaks in the input can change the output by large amounts - very high is not good.
- If the condition number is infinite then the matrix is singular (determinant is 0) -  the 2D space has flattened to a 1D line post that matrix transformation - matrix is singular and has linearly dependent rows or columns

***
In data science, 
* the condition number tells you how sensitive a linear regression model's outputs (the coefficient weights) are to small changes in the inputs (the feature data). 
* A very high condition number means your matrix (XT X)^^-1 part of the beta matrix is close to becoming singular (determinant near 0), indicating severe multicollinearity.
* Infinite condition Number tells that the matrix has linearly dependent rows and columns and suffer from perfect multicollinearity.

In predictive modeling and regression, a very high condition number is an immediate red flag for severe multicollinearity (independent variables are highly correlated), 
meaning your model's coefficients are highly unstable as they suffer from high standard deviations even though the estimates are unbiased. 

***
**Condition Number = ||A||.||A^-1||**
**or, Condition Number = max(singular values)/ min(singular values) for a generic (m,n) shape matrix**

Here ||A|| is called the norm of the matrix. By default numpy uses the spectral/2D norm of the matrix (np.linalg.matrix_norm())
One common norm is also the Frobenius norm which is the sqrt(sum of square of matrix elements) - (np.linalg.matrix_norm(array,'fro'))

Using the spectral norm we,
* perfrom singular value decomposition on the matrix
* calculate the maximum and minimum singular values
* calcualte Condition Number using their ratio

***
* If you have your data matrix of mXn shape then use **Ratio of largest to smallest singular value**.
* If you have the symmetric matrix (mirrored across diagonal) like the covariance matrix then use **sqrt(Ratio of largest to smallest eigen value)** 

```python
np.random.seed(42) #for reproducibility
r1 = np.random.randint(50,80,(3,6))
print(f"Condition number of the matrix \n{r1}\n is {np.linalg.cond(r1)}")
```

##### 2. Rank of a Matrix

***
Rank of a matrix is a number that tells how many linearly independent rows or columns are present in a matrix. If a row or column is formed from another row(s) or column(s) using mathematical operations like addition, subtraction, multiplication then that row or column is linearly dependent i.e. the row or column can be written in a linear combination of other row(s) or column(s).

***
- Rank can never exceed the smallest dimension of your matrix meaning that the maximum possible rank for a (3,5) matrix is 3. Rank <= min(m,n)
- In case there are no linearly dependent rows or columns, the matrix assumes full rank which means no row or column can be written in a linear combination of other rows or columns and the matrix is linearly independent. 
- In case the matrix is singular (det =0) the matrix doesn't necessarily have rank =0. However the rank is strictly less than the smallest dimension of the matrix

***

To detect singularity of a matrix we use the concept of **Condition Number**.

Instead of relying on rank alone, we should check the Condition Number. 
A matrix doesn't instantly lose rank if it becomes more singular. 
-	A matrix becomes ill-conditioned first which means the condition number keep on increasing. 
-	A high condition number means your matrix is dangerously close to losing rank. 


```python
a = np.random.randint(20,40,size = (3,5))
print(a)
print("\n")

# inititalize a square matrix
sq_matrix = np.random.randint(10,50, size = (4,4))

# rank of a matrix - no of linearly independent rows or columns of a matrix
	# if a row or column is formed by addition subtraction multiplication of one or more rows of the matrix - then that row is linearly dependent
	# the rank can never be larger than the smallest dimension of your matrix i.e. Rank<= min(m,n)
		# For a (3 rows, 5 columns), the maximum possible rank is 3.

#linearly independent matrix
print(f"Rank of a Matrix \n{a}\n is")
print(np.linalg.matrix_rank(a))
print("which shows it is linearly independent")
print("\n")

#linearly dependent matrix
print(f"Rank of a Matrix \n{np.array([[1,2,3,4],[4,8,12,16]])}\n is")
print(np.linalg.matrix_rank(np.array([[1,2,3,4],[4,8,12,16]])))
print("which shows it is not linearly independent")
print("\n")
```

##### 4 Deteminant of a Matrix - only for a square matrix

**Basis Vectors**

A basis vector is a vector which you can use to span every point or vector in the space (2D or 3D or nD) that has them. Basically you can go to any desired scalar or vector coordinate space with the help of those vectors. 

#### Unit Vectors in Different Dimensions

##### In 2D Space (R^2 space)

The standard unit vectors in 2-dimensional space are
- i = [1,0]
- j = [0,1]

These are unit vectors or basis vectors and they span the entire 2D space R^2
These represent the unit vectors along the **x-axis** and **y-axis** respectively.

Using these vectors you can reach to any point in the 2D coordinate space. 


#### In 3D Space (R^3 space)

The standard unit vectors in 3-dimensional space are:
- i = [1,0,0]
- j = [0,1,0]
- k = [0,0,1]

Here, *k* vector is added to represent the unit vector along the **z-axis**.
These are unit vectors or basis vectors and they span the entire 2D space R^3

---

**What is determinant actually?**

For a standard 2D matrix the determinant is basically the area of the parallelogram that is created when you apply the matrix transformation i.e., a linear transformation on the basis vectors i and j. This transformation changes or stretches the entire 2D space, not specifically a particular point in the co-ordinate space. 

Imagine pulling the 2D or 3D space from one particular side but the origin is nailed to be fixed. Also you can stretch it in such a way that all the gridlines are parallel to each other and they are still straight lines. 

In case of a 2D space,  i = [1,0] and j = [0,1]. 

- Imagine this to be a unit square on a 2D grid at the origin where the sides of the square represents the unit basis vectors. Using this you can go to any point within the 2D grid. 
- Now, suppose you apply a matrix transformation to the entire grid. Imagine pulling this from a side so that all the gridlines are now stretched to a particular direction but all are still straight lines and parallel to each others. 
- The matrix moves the basis vectors i and j to new positions, turning our perfect grid into a slanted, stretched grid. 
- The original 1 X 1 square is now a skewed parallelogram.
- The Determinant is the area of that new parallelogram.

* In 2D: The determinant is the area of the transformed unit square (parallelogram).
* In 3D: The determinant is the volume of the transformed unit cube (parallelepiped).
* In Higher Dimensions ($nD$): It is the factor by which the "$n$-dimensional volume" changes.

***
If determinant of a matrix is +6 no matter what the shape of the matrix is, it will always scale the area by a factor of +6
This means that if you pass a (2,3) matrix or a (7,9) matrix if the determinant is 6 then both of those matrix transformations will scale the area of the unit square created by the basis vectors by +6 or a factor of +6 -  new area of the parallelogram created. 

Two matrices can have completely different entries, pull space in different directions, and create entirely different looking shapes. 
But if their determinants are equal, the total amount of space they "create" or expand is exactly the same.
***

#### SVD from scratch

A matrix is a complex linear transformation that changes the entire coordinate space. It rotates, transforms, flips or does a combination of those to the space the point or vector lies in the coordinate space. Imagine if you apply a matrix A on every point on the circle it stretches, flips, rotates, changes dimension of that circle and transforms it into an ellipse (say). The question was can any complex transformation be broken down into simple ones?

A = $U$ * $\Sigma$ * $V^T$

This shows that any (m,n) matrix can be essentially broken down into a series of linear transformation ($V^T$,$\Sigma$,$U$)

In SVD we decompose the matrix into three components namely,
- $U$: Left singular matrix that is created by getting the Eigen Vectors for AAT matrix
- $\Sigma$: Sigma Matrix  also a diagonal matrix consiting of the singular values which are sqrt( eigen values of AAT or ATA matrix) arranged in a descending order
- $V$: Right singular matrix that is created by getting the Eigen Vectors for ATA matrix

***
**The magic of SVD is that it finds a special set of perpendicular/ orthogonal vectors in the input space ($V$) that map to a strictly perpendicular set of vectors in the output space ($U$).**

**How does this work Geometrically ?**

- $V$ is an orthogonal matrix, meaning its columns form an orthonormal basis - new perpendicular axes (the right singular vectors). Multiplying by $V^T$ changes the basis of your input vector $\vec{x}$ from the standard Cartesian coordinates to this new coordinate system. It preserves lengths and angles; it is purely a rotation.

- $\Sigma$ is a diagonal matrix containing the singular values ($\sigma_1, \sigma_2, \dots$). This step independently scales each component of the freshly rotated vector. If a singular value is large, the vector stretches along that axis; if it's zero, that dimension is completely collapsed.

- $U$ is another orthogonal matrix -new axes that are 90 degrees in the final output vector space (the left singular vectors). It takes the scaled vector and applies a second, independent rotation. This lands the vector in its final home—the output vector space.

***
- V - gives the direction of the principal axes that are orthogonal to each other in the original input vector space. 
- VT - gives the location or point of the original vector when we use the columns in V as axis - keeps the length angle and other properties of the vector intact. 
- E - gives the scalar by which the vector is scaled in the new vector space - when we use the columns in V as axis
- U - gives another rotation that gives back the location of the scaled vector in the final output vector space where also the axes are orthogonal to each other

These are the new coordinates of the vector in the newly crerated vector space. 

***
U and V are **Orthogonal Matrices** or **Orthonormal Matrices**. This means that the vectors that create the U matrix or V matrix are pependicular to each other and their dot product is 0. 

```python
import numpy as np
#np.random.seed(0)
A = np.random.randint(10,20, (3,5)) 
print(A)
print("\n")

# SVD (A) = U @ E @ VT

#computing ATA
ATA = (np.transpose(A)) @ A
print(f"ATA is \n {ATA}")
print("\n")

# Compute the Eigenvalues and Eigenvectors of ATA
eigval, eigvec = np.linalg.eig(ATA)
print(f"Eigenvalues are {eigval.round(decimals=3)}\n") # square root of these values gives the singular values of E matrix
print(f"Eigenvectors are {eigvec.round(decimals=3)}\n") # this is the V matrix

# Create the VT matrix
VT = np.transpose (eigvec.round(decimals=3))
print(f"The VT matrix is \n{VT}\n")
print(f"The shape of the VT matrix is {VT.shape}\n")

#Create the E matrix
sqrt_eigval = np.sqrt(eigval.round(decimals=3))
print(f"The singular values across diagonals are {sqrt_eigval}\n")

#compute the diagonal matrix
E= np.diag(sqrt_eigval.round(decimals=3))
print(f"The sigma(E) matrix is {E} and the shape is {E.shape}\n") 

# Compute U = A @ V @ E^-1

#E_inv = np.linalg.inv(E.round(decimals=3))  

# U = (A @ V @ E_inv).round(decimals=3)
# print(f"The Matrix U is {U} and the shape is {U.shape}")

```

#### Eigenvalue Decomposition

Eigen value decomposition, a special case of Singular Value Decomposition, is also a matrix factorization technique but works only for square matrices. It decomposes the matrix into three components namely

A = Q @ V @ Q^-1

***
When you apply the matrix transformation on a vector it essentially changes the space in which the vector lies by rotation stretching, flipping or a combination of them. But there are some magical vectors that do not rotate, instead they are just scaled in size. These vectors are called eigen vectors and the amount by which they are scaled are called eigen values. 
***
What happens in eigen decomposition is that you decompose or factor the matrix into three different components. 
Q - this represents the eigenvector matrix where each column represents each eigen vector of the matrix as per each characteristic root
V - this represents the eigenvalue matrix or lambdas which are also the characteristic roots for matrix A in descending order of magnitude - diagonal matrix
Q^-1 - this represents the inverse of the eigenvector matrix Q

**Intuition**
* Say there is a vector X = [x1,x2]. Now when you apply a matrix transformation the Q^-1 @ X transforms the standard basis of the vector into eigenvector basis. This means 
if X=[(3,0),(0,2)] then it moves 3 units on the X axis and 2 units on the Y axis to go to the point. 

* When we do Q^-1 @ X it changes to eigenvector basis. That means if Q^-1 @ X = [q1,q2] then if we move along q1 units along eigenvector1 and then q2 units along eigenvector 2 then we will be able to reach that vector in the eigenvector space. 

* V @ Q^-1 @ X scales that original vector in the space where the basis are eigenvector basis. 

* Q @ V @ Q^-1 @ X re-changes the basis to standard basis space like 2D space and we get the scaled vector post matrix transformation in the same 2D space back. 

This is extremely useful during dimensionality reduction where we use PCA to reduce the number of unnecessary features and we apply Eigendecompostion on the Variance-Covariance matrix and we get the columns of V(eigen value) matrix gives the amount of variance captured by the eigenvectors in matrix Q.

##### Element-wise Boundaries & Trimming
***
**np.clip(array, a_max, a_min) solves the problem of Winsorization**
- Restricts an array to user defined maximum and minimum
- Anything outside the defined range gets rounded to the max and min value
- If you don't want to trim one particular edge put a_min = None or a_max = None

```python
age_data = np.array([12, 25, 34, 99, -5, 45])
display(age_data)

print(np.clip(age_data, a_min= 18, a_max=80))
print("\n")

print(np.clip(age_data, a_min=None,a_max=80))
```
