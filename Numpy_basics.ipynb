##### 7. Creating evenly spaced arrays over a specified interval
Use np.linspace(start=,stop=,num=,endpoint=,retstep=,dtype=)

* start = start of the sequence
* stop = stop value of the sequence
* num = number of samples - by default it it 50
* endpoint = True/False (boolean) - if True it ends at the stop value if False then the stop value is excluded
* retstep - Step used to create the matrix

***
**Difference between np.arrange() and np.linspace() is**

using np.linspace()
- we can specify how many samples we want - by defualt 50 samples
- we can end the array at the stop value if we want

using np.arange()
- we can specify the step size we want 



```python

#Endpoint =True - stops ar 50/last value will be 50 included
print(np.linspace(20,50,endpoint=True,num=30, dtype='int'))

#Endpoint =False hence stops at 49 and num not specified hence 50 samples by default
print(np.linspace(20,50,endpoint=False, dtype='int'))
```

##### Boolean Indexing

```python
np.random.seed(42)
# 1D Array
array_1D = np.random.randint(30,100,20)
print(array_1D,"\n")
print("\n")

print("Print all elements which are greater than 50")
print(array_1D[array_1D>50])
print("\n")

print("Print the element 89")
print(array_1D[array_1D == 89])
print("\n")

print("Print the numbers between 45 and 75") 
print(array_1D[(array_1D>=45) & (array_1D<=75)])
print("\n")

print("Print the numbers less than 40 or greater than 80") 
print(array_1D[(array_1D <=40) | (array_1D >= 80)])
print("\n")

print("Print the numbers not less than 40 ") 
print(array_1D[~(array_1D <=40)])
print("\n")
```

```python
np.random.seed(42)
# 2D Array
array_2D = np.random.randint(30,100,(4,5))
print(array_2D,"\n")
print("\n")

print("Print the element that are greater than 50")
print(array_2D[array_2D>50]) #returns a 1D array
print("\n")

print("Print the element that are greater than 50 but less than 75")
print(array_2D[(array_2D>50) & (array_2D<=75)]) #returns a 1D array
print("\n")

# By default boolean indexing occurs row-wise for Numpy
print("Print the rows where the third column elements are greater than 50 but less than 75")
print(array_2D[array_2D[:,2]>85]) #returns all the rows that has column 3 values gt 60
print("\n")

print("Print the rows where the third column elements are greater than 85 and less than 80")
print(array_2D[(array_2D[:,2]>85) & (array_2D[:,0]<80)])
print("\n")
```

##### 3. Deteminant of a Matrix - only for a square matrix

**Basis Vectors**

A basis vector is a vector which you can use to span every point or vector in the space that has them. Basically you can go to any desired scalar or vector coordinate space with the help of those vectors. 

#### Unit Vectors in Different Dimensions

##### In 2D Space (R^2 space)

The standard unit vectors in 2-dimensional space are
- i = [1,0]
- j = [0,1]

These are unit vectors or basis vectors and they span the entire 2D space R^2
These represent the unit vectors along the **x-axis** and **y-axis** respectively.


#### In 3D Space (R^3 space)

The standard unit vectors in 3-dimensional space are:
- i = [1,0,0]
- j = [0,1,0]
- k = [0,0,1]

Here, *k* vector is added to represent the unit vector along the **z-axis**.
These are unit vectors or basis vectors and they span the entire 2D space R^3

---

**What is determinant actually?**

For a standard 2D matrix the determinant is basically the area of the parallelogram that is created when you apply the matrix transformation i.e., a linear transformation on the basis vectors i and j. This transformation changes or stretches the entire 2D space, not specifically the a particular point or co-ordinate space. 

In case of a 2D space,  i = [1,0] and j = [0,1]. 

- Imagine this to be a unit square on a 2D grid at the origin where the sides of the square represents the unit basis vectors. Using this you can go to any point within the 2D grid. 
- Now, suppose you apply a matrix transformation to the entire grid. Imagine pulling this from a side so that all the gridlines are now stretched to a particular direction but all are still straight lines and parallel to each others. 
- The matrix moves the basis vectors i and j to new positions, turning our perfect grid into a slanted, stretched grid. 
- The original 1 X 1 square is now a skewed parallelogram.
- The Determinant is the area of that new parallelogram.

* In 2D: The determinant is the area of the transformed unit square (parallelogram).
* In 3D: The determinant is the volume of the transformed unit cube (parallelepiped).
* In Higher Dimensions ($nD$): It is the factor by which the "$n$-dimensional volume" changes.

***
If determinant of a matrix is +6 no matter what the shape of the matrix is, it will always scale the area by a factor of +6
This means that if you pass a (2,3) matrix or a (7,9) matrix if the determinant is 6 then both of those matrix transformations will scale the area of the parallelogram created by the basis vectors by +6 or a factor of +6

Two matrices can have completely different entries, pull space in different directions, and create entirely different looking shapes. 
But if their determinants are equal, the total amount of space they "create" or expand is exactly the same.
***

```python
display(sq_matrix)
# determinant of a matrix - occurs only for a square matrix - multiplies the matrix several times raised to the power n
print("The determinant is")
print(np.linalg.det(sq_matrix))
```

##### 4. Eigen Values and Eigen Vectors - only for a square matrix

| Function | Matrix Type | What it Returns | Speed / Stability |
| :--- | :--- | :--- | :--- |
| **np.linalg.eig** | General Square | Eigenvalues + Eigenvectors | Standard |
| **np.linalg.eigvals** | General Square | Eigenvalues only | Faster (skips vectors) |
| **np.linalg.eigh** | Symmetric / Hermitian | Eigenvalues + Eigenvectors | Fastest & Most Stable |
| **np.linalg.eigvalsh** | Symmetric / Hermitian | Eigenvalues only | Blazing Fast |

1. If A = AT then it is a symmetric matrix comprising of real numbers

2. A matrix is Hermitian if it equals its own conjugate transpose ($A = A^H$). This means you flip the rows and columns and change the sign of the imaginary parts (e.g., $2 + 3i$ becomes $2 - 3i$). The diagonal has to be a real number though for this matrix.

```python
print(sq_matrix)
print("\n")

eigenvalues, eigenvectors = np.linalg.eig(sq_matrix)

# eigen values
print("Eigenvalues: ", eigenvalues)
print("\n")

# eigen vectors
print("Eigenvectors: ", eigenvectors)
```

##### 5. Condition Number of a Matrix

***
The condition number measures how sensitive a system of linear equations Ax=B (imagine linear regression) is to small changes or numerical rounding errors. Basically assess the numerical stability.
- If the condition number is low then it is good, small changes in the data does not change the output by much - close to 1 is good as there sis computation statbility.
- If the condition number is high then it is a ill-conditioned matrix, small changes or tweaks in the data can change the output by large amounts - very high is not good.
- If the condition number is infinite then the matrix is singular (determinant is 0) -  the 2D space has flattened to a 1D line post that matrix transformation - matrix is singular and has linearly dependent rows or columns

***
In data science, 
* the condition number tells you how sensitive a linear regression model's outputs (the coefficient weights) are to small changes in the inputs (the feature data). 
* A very high condition number means your matrix (XT X)^^-1 part of the beta matrix is close to becoming singular (determinant near 0), indicating severe multicollinearity.
* Infinite condition Number tells that the matrix has linearly dependent rows and columns and suffer from perfect multicollinearity.

In predictive modeling and regression, a very high condition number is an immediate red flag for severe multicollinearity (independent variables are highly correlated), 
meaning your model's coefficients are highly unstable as they suffer from high standard deviations even though the estimates are unbiased. 

***
**Condition Number = ||A||.||A^-1||**
or, **Condition Number = max(singular values)/ min(singular values)**

Here ||A|| is called the norm of the matrix. By default numpy uses the spectral/2D norm of the matrix (np.linalg.matrix_norm())
One common norm is also the Frobenius norm which is the sqrt(sum of square of matrix elements) - (np.linalg.matrix_norm(array,'fro'))

Using the spectral norm we,
* perfrom singular value decomposition on the matrix
* calculate the maximum and minimum singular values
* calcualte Condition Number using their ratio

***
* If you have your data matrix of mXn shape then use **Ratio of largest to smallest singular value**.
* If you have the scaled version of covariance matrix (square matrix) then use **sqrt(Ratio of largest to smallest eigen value)** 

```python
np.random.seed(42) #for reproducibility
r1 = np.random.randint(50,80,(3,6))
print(f"Condition number of the matrix \n{r1}\n is {np.linalg.cond(r1)}")
```
